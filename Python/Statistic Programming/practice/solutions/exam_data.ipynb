```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/kinpkt/catlover-cs131-practice/refs/heads/main/practice_dataset/exam_data.csv')
df
```

**Part 1 : หาอายุเฉลี่ยของนักเรียนในสนามสอบนี้**

```python
age_mean = np.mean(df.age)
print(f'Mean age for students = {age_mean}')
```

**Part 2 : หา S.D. ของเกรดเฉลี่ยนักเรียนในสนามสอบนี้**

```python
# โจทย์ข้อนี้อาจตีความได้สองแบบ (คนคิดโจทย์ error เอง; อาจใช้สูตร S.D. ของประชากรหรือของตัวอย่างก็ได้)
sd_population = np.std(df.gpax)
sd_sample = np.std(df.gpax, ddof=1)

print(f'Population S.D. = {sd_population}\nSample S.D = {sd_sample}')
```

**Part 3 : หาวิชาที่นักเรียนที่สมัครมากและน้อยที่สุดและจำนวนนักเรียนที่สมัครวิชานั้น**

```python
SUBJECTS = ['Math 1', 'Math 2', 'Science', 'Physics', 'Social Studies', 'English', 'Chinese', 'Japanese']

math1_count = sum(df.math1)
math2_count = sum(df.math2)
sci_count = sum(df.sci)
phy_count = sum(df.phy)
soc_count = sum(df.soc)
eng_count = sum(df.eng)
chn_count = sum(df.chn)
jpn_count = sum(df.jpn)

count_list = [math1_count, math2_count, sci_count, phy_count, soc_count, eng_count, chn_count, jpn_count]
max_count = max(count_list)
min_count = min(count_list)
max_subject = SUBJECTS[count_list.index(max_count)]
min_subject = SUBJECTS[count_list.index(min_count)]
print(f'The most registered subject is {max_subject} with {max_count} students.')
print(f'The least registered subject is {min_subject} with {min_count} students.')
```

**Part 6 : เปรียบเทียบจำนวนนักเรียนที่สมัครสอบ Math 1 และ Math 2**

```python
print(f'Math 1 students are {"more than" if math1_count > math2_count else "less than"} than Math 2 students.')
```
