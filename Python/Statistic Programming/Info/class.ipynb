```python
df_ct = pd.crosstab(titanic_data.Survived, titanic_data.Embarked,margins=False)
print(type(df_ct))
df_ct
```

```python
df_ct = pd.crosstab(titanic_data.Pclass, titanic_data.Survived , margins= False )
df_ct
```

```python
surv_div_death = df_ct[1]/df_ct[0] # cuz หัวข้อคือ int
surv_div_death
```
